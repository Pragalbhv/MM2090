```python
from sage.misc.prandom import randrange
from sage.plot.point import point
from sage.plot.circle import circle
```

```python
x=[randrange(100) for i in range(25)]
y=[randrange(100) for i in range(25)]
n=zip(x,y)
print(n)
p1=points(n,size=12,color='green')
p1.show()
```

```python
m=randrange(24)
o= (x[m],y[m])
print(o)
```

```python
var('diff','d')
g=Graphics()
m=Graphics()
d = 50
for i in  range(25):
    diff=(((x[i])-o[0])^2+((y[i]-o[1])^2))^(0.5)
    if diff <=d:
        g = g+(points((x[i],y[i]),size=25,color='red'))
    if diff >d:
            m=m+(points((x[i],y[i]),size=25,color='green')) 
```

```python
p2=points(o,size=75,color='blue')
c=circle((o[0],o[1]), d)
(p2+g+m+c).show()
```
