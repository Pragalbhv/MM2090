Vornoi Daigram:-
In mathematics, a Voronoi diagram is a partition of a plane into regions close to each of a given set of objects. In the simplest case, these objects are just finitely many points in the plane

The Delaunay triangulation is a triangulation which is equivalent to the nerve of the cells in a Voronoi diagram, i.e., that triangulation of the convex hull of the points in the diagram in which every circumcircle of a triangle is an empty circle

```octave
x = gallery('uniformdata',[1 10],0);
y = gallery('uniformdata',[1 10],1);
tri = delaunay (x, y);
triplot (tri, x, y,'k');
hold on;
plot (x, y, "r*");
axis ([0,1,0,1]);
voronoi(x,y)
axis equal
```
